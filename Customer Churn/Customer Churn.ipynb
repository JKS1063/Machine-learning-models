```python
plt.figure(figsize = (8,6))
sns.boxplot(x = data.Contract, y = data.tenure)
plt.xlabel('Contract', fontweight = 'bold')
plt.ylabel('Tenure', fontweight = 'bold')
plt.title('Tenure & Contract', fontweight = 'bold')
plt.grid(True)
plt.show()
```

* Two year contract custmers are having highest tenure with company followed by One-year and month-to-month.
* Two year contract: customers who opted for this contract have the least tenure time of 15 months and most being as around 75 months with the company. Customers who are opting this contract have an average of 63 months as tenure time with company.
* One year contract: customers who opted for this contract have the least tenure time of one month and most being as around 75 months with the company. Customers who are opting this contract have an average of 44 months as tenure time with company.
* Month-to-Month contract: customers who opted for this contract have the least tenure time of 1 month and most being as around 68 months with the company. Customers who are opting this contract have an average of 12 months as tenure time with company.
* We can say that most of the old customers having hith tenure time are preferring Two year contract, followed by One year contract.
* Most of the new customers having least tenure time with company are opting for month-to-month contract.

### 3. Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
HeatmapLabels = ['True', 'False']
```

#### Linear Regression Model

```python
x1 = df[['tenure']]
y1 = df.MonthlyCharges
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1,
                                                        test_size = 0.3,
                                                        random_state = 42)
```

```python
lir = LinearRegression()
```

```python
lir.fit(x1_train, y1_train)
```

```python
lir_predict = lir.predict(x1_test)
```

```python
print('R-sq value:', round(metrics.r2_score(y1_test, lir_predict), 2))
```

```python
error = round(np.sqrt(metrics.mean_squared_error(y1_test, lir_predict)), 2)
error
```

```python
sns.distplot(y1_test, label = 'Actual')
sns.distplot(lir_predict, label = 'Predicted')
plt.legend()
plt.show()
```

#### Logistic Regression Model

```python
x2 = df[['MonthlyCharges']]
y2 = df.Churn
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, 
                                                        test_size = 0.35,
                                                        random_state = 42,
                                                        shuffle = True)
```

```python
lor = LogisticRegression()
```

```python
lor.fit(x2_train, y2_train)
```

```python
lor_predict = lor.predict(x2_test)
```

```python
print('Logistic Regression Accuracy Score:',round(metrics.accuracy_score(y2_test, lor_predict), 2))
```

```python
cm = metrics.confusion_matrix(y2_test, lor_predict)

sns.heatmap(cm, annot = True, vmin = 0, vmax = 400, fmt = '.2f',
            xticklabels=HeatmapLabels, yticklabels=HeatmapLabels)
```

```python
sns.distplot(y2_test, label = 'Actual')
sns.distplot(lor_predict, label = 'Predicted')
plt.legend()
plt.show()
```

```python
print(metrics.classification_report(y2_test, lor_predict))
```

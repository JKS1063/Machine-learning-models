#### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.airline = le.fit_transform(df.airline)
df.source_city = le.fit_transform(df.source_city)
df.departure_time = le.fit_transform(df.departure_time)
df.stops = le.fit_transform(df.stops)
df.arrival_time = le.fit_transform(df.arrival_time)
df.destination_city = le.fit_transform(df.destination_city)
df['class'] = le.fit_transform(df['class'])
```

#### Feature Selection

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm', fmt = '.1f')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []

for col in df.columns:
    if ((df[col].dtype != 'object') & (col != 'price')):
        col_list.append(col)
```

```python
x = df[col_list]
vif_data = pd.DataFrame()
vif_data['feature'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i)
                  for i in range(len(x.columns))]

print(vif_data)
```

```python
x = df.drop(['price', 'flight'], axis = 1)
y = df.price
```

#### Splitting the data into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,
                                                  random_state = 42)
```

#### Standardizing the Model

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)

y_pred = np.round(lr.predict(x_test), 2)

diff = pd.DataFrame(np.c_[y_test, y_pred],
                    columns = ['Actual', 'Predicted'])

diff
```

#### R2_score, Mean Absolute Error, Mean Score Error
#### Mean Absoulte Percentage Error, Root Mean Square Error

```python
from sklearn import metrics

print('R-Square value should be more than 65%, Lower the RMSE and MAPE better the model')

print('R-Square value is: ', metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error is: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Score Error is: ', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Percentage Error is: ', metrics.mean_absolute_percentage_error(y_test, y_pred))
print('Root Mean Square Error is: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
sns.distplot(y_test, label = 'Actual')
sns.distplot(y_pred, label = 'Predicted')
plt.legend()
```

#### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(x_train, y_train)

y_pred = dt.predict(x_test)

from sklearn import metrics

print('R-Square value should be more than 65%, Lower the RMSE and MAPE better the model')

print('R-Square value is: ', metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error is: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Score Error is: ', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Percentage Error is: ', metrics.mean_absolute_percentage_error(y_test, y_pred))
print('Root Mean Square Error is: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
sns.distplot(y_test, label = 'Actual')
sns.distplot(y_pred, label = 'Predicted')
plt.legend()
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rfc = RandomForestRegressor()

rfc.fit(x_train, y_train)

y_pred = rfc.predict(x_test)

from sklearn import metrics

print('R-Square value should be more than 65%, Lower the RMSE and MAPE better the model')

print('R-Square value is: ', metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error is: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Score Error is: ', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Percentage Error is: ', metrics.mean_absolute_percentage_error(y_test, y_pred))
print('Root Mean Square Error is: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
sns.distplot(y_test, label = 'Actual')
sns.distplot(y_pred, label = 'Predicted')
plt.legend()
```
